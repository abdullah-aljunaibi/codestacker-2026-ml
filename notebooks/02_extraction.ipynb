# Receipt Field Extraction Walkthrough

This notebook exercises the production extraction pipeline in `src/extractor.py` against the bundled sample data.

```python
from __future__ import annotations

import sys
from pathlib import Path

import pandas as pd
from PIL import Image

ROOT = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd().resolve()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from src.extractor import extract_date, extract_fields, extract_text, extract_total, extract_vendor
```

```python
sample_image = ROOT / "dummy_data" / "train" / "images" / "r001.png"
sample_image
```

```python
Image.open(sample_image)
```

```python
ocr_text = extract_text(str(sample_image))
print(ocr_text)
```

```python
parsed_from_text = {
    "vendor": extract_vendor(ocr_text),
    "date": extract_date(ocr_text),
    "total": extract_total(ocr_text),
}
parsed_from_text
```

```python
pipeline_result = extract_fields(str(sample_image))
pipeline_result
```

```python
sample_paths = sorted((ROOT / "dummy_data" / "train" / "images").glob("*.png"))[:5]
preview_rows = []
for path in sample_paths:
    result = extract_fields(str(path))
    preview_rows.append(
        {
            "image": path.name,
            "vendor": result["vendor"],
            "date": result["date"],
            "total": result["total"],
        }
    )

pd.DataFrame(preview_rows)
```

## Notes

- `extract_fields` is the preferred entry point because it keeps OCR text and word-level structure aligned.
- Vendor extraction uses the specialized helper in `src/extractors/vendor.py`.
- Date and total extraction rely on regex patterns with normalization fallbacks.
